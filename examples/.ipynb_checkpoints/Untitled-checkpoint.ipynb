```python
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
import seaborn as sns

df = pandas.read_csv('../data/players.csv')
```

```python
sns.set(color_codes=True)
qbs = df[df['position'] == 'QB'][["passing_yds", "passing_tds", "passing_att", "own_score", "opp_score"]].fillna(0)
qbs["score_diff"] = qbs["own_score"] - qbs["opp_score"]
del qbs["own_score"]
del qbs["opp_score"]
qb_corr = qbs.corr()

g = sns.heatmap(qb_corr)
```

```python
sns.set(color_codes=True)
rbs = df[df['position'] == 'RB'][["rushing_yds", "rushing_tds", "rushing_att", "own_score", "opp_score"]].fillna(0)
rbs["score_diff"] = rbs["own_score"] - rbs["opp_score"]
rbs[""]
del rbs["own_score"]
del rbs["opp_score"]
rb_corr = rbs.corr()
sns.heatmap(rb_corr)
```
