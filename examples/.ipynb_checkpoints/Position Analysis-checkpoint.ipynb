```python
# Loading default settings

%matplotlib inline
import numpy as np
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

df = pandas.read_csv('../data/players.csv')
df['playing_against'] = le.fit_transform(df['playing_against'])
```

```python
qbs = df[df["position"] == "QB"]
sns.set(color_codes=True)
qb_stats = qbs[["passing_yds", "passing_tds", "passing_att","is_home", "std_score", "playing_against"]].fillna(0)

qb_stats_corr = qb_stats.corr()

g = sns.heatmap(qb_stats_corr, annot=True)
```

```python
rbs = df[df["position"] == "RB"]
sns.set(color_codes=True)
rb_stats = rbs[["rushing_yds", "rushing_tds", "rushing_att","is_home", "std_score", "playing_against"]].fillna(0)

rb_stats_corr = rb_stats.corr()

g = sns.heatmap(rb_stats_corr, annot=True)
```
