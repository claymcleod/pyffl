```python
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
import seaborn as sns

df = pandas.read_csv('../data/players.csv')
df["score_diff"] = df["own_score"] - df["opp_score"]
```

```python
sns.set(color_codes=True)
qbs = df[df['position'] == 'QB'][["passing_yds", "passing_tds", "passing_att", "score_diff"]].fillna(0)
qb_corr = qbs.corr()

g = sns.heatmap(qb_corr)
```

```python
sns.set(color_codes=True)
rbs = df[df['position'] == 'RB'][["rushing_yds", "rushing_tds", "rushing_att", "score_diff"]].fillna(0)
rb_corr = rbs.corr()
sns.heatmap(rb_corr)
```

```python
sns.set(color_codes=True)
win = df[["own_team_rushing_yds", "own_team_passing_yds", "own_team_pos_time",  "score_diff"]]
win["ratio_pass_to_run"] = df["own_team_passing_yds"] / df["own_team_rushing_yds"]
win_corr = win.corr()
sns.heatmap(win_corr, annot=True)
```

```python
team = df[df["team"] == "SEA"]
ax = sns.kdeplot(team["own_score"], team["opp_score"],
                 cmap="Reds", shade=True, shade_lowest=False)
```
